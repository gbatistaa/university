```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
mama = imread('mamografia.png')
```

```python
plt.imshow(mama,cmap='gray')
```

```python
h,_ = np.histogram(mama,bins=256)
```

```python
plt.plot(h,'-k')
```

```python
aeroporto = imread('aeroporto.png')
```

```python
h,_ = np.histogram (aeroporto,256)
plt.plot (h,'-k')
```

```python
plt.imshow(aeroporto,cmap='gray')
```

```python
mama2 = 255 - mama
plt.imshow(mama2,cmap='gray')
```

```python
h,_ = np.histogram (mama2,256)
plt.plot (h,'-k')
```

```python
carro = imread ('HI00008918.png',as_gray=True)
```

```python
carro = carro * 255
carro = carro.astype ('int')
```

```python
plt.imshow (carro,cmap='gray')
```

```python
h,_ = np.histogram (carro,256)
plt.plot (h,'-k')
```

```python
carro2 = carro*2.5
carro2[carro2>250] = 255
carro2 = carro2.astype('int')
```

```python
h,_ = np.histogram (carro2,256)
plt.plot (h,'-k')
```

```python
plt.imshow(carro2,cmap='gray')
```

A operação anterior realizou o espalhamento do histograma nas regiões onde havia mais massa e, a compressão, nas regiões onde havia pouca. Isto  produziu o realce de contraste desejado. Há, no entanto, uma função que tem o mesmo efeito e não requer a inspeção manual do histograma. Esta é a função de probabilidade acumulada e é obtida pela equalização do histograma.
