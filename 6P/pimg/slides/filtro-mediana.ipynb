```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.filters.rank import median
from skimage.morphology import disk
from scipy.signal import convolve2d
```

## Filtro da mediana

```python
img = imread ('lena.png')
```

```python
img_ruido = img.copy()
img_ruido [np.random.random(img.shape)<0.1] = 0  # ruido do tipo pimenta (10%)
img_ruido [np.random.random(img.shape)>.9] = 255 # ruido do tipo sal (10%)
```

```python
plt.imshow (img_ruido,cmap='gray')
```

```python
img_med = median (img_ruido,footprint=np.ones((13,13)).astype('uint8'))
```

```python
_,ax = plt.subplots (1,2)
ax[0].imshow (img,cmap='gray')
ax[1].imshow (img_med,cmap='gray')
```

## Gradiente de Sobel

```python
mx = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
my = mx.T
```

```python
img_grad = np.abs (convolve2d(img,mx,mode='same')) + np.abs (convolve2d(img,my,mode='same'))
```

```python
plt.imshow(img_grad,cmap='gray')
```
