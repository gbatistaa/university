# Quest√£o 1 - 

```python
car = imread("HI00008918.png")
car_norm = car / 255.0
```

```python
plt.imshow(car_norm, cmap="gray")
```

```python
h,_ = np.histogram(car_norm, bins = 256)
plt.plot(h, "-k")
```

```python
gamma1 = 0.2
gamma2 = 5
```

```python
car_copy1 = car_norm.copy().astype("float")
car_copy_gamma1 = np.power(car_copy1, gamma1)

car_copy_gamma1[car_copy_gamma1 > 255] = 255

plt.imshow(car_copy_gamma1, cmap="gray")

```

```python
h,_ = np.histogram(car_copy_gamma1, bins = 256)
plt.plot(h, "-k")
```

```python
car_copy2 = car_norm.copy().astype("float")
car_copy_gamma2 = np.power(car_copy2, gamma2)

car_copy_gamma1[car_copy_gamma1 > 255] = 255

plt.imshow(car_copy_gamma2, cmap="gray")
```

```python
h,_ = np.histogram(car_copy_gamma2, bins = 256)
plt.plot(h, "-k")
```

# Quest√£o 2 - Equaliza√ß√£o Histogr√¢mica

```python
lena_img = imread("lena.png")
lena_img = lena_img.astype("int")
plt.imshow(lena_img, cmap="gray")
```

### Histograma - Lena Padr√£o

```python
h,_ = np.histogram(lena_img, bins=256)
plt.plot(h, "-k")
```

### Equaliza√ß√£o Histogr√¢mica

```python
cdf  = h.cumsum()
cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
cdf_norm = cdf_norm.astype("int")
lena_img_equalized = cdf_norm[lena_img]
cdf_norm = cdf_norm.astype('int')
```

### Plotagem equalizada

```python
plt.imshow(lena_img_equalized, cmap="gray")
```

### Histograma Equalizado

```python
e,_ = np.histogram(lena_img_equalized, bins=256)
plt.plot(e, "-k")
```
