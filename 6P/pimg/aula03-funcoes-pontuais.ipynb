## Realce de contraste
### Multiplicação por um valor constante

```python
placa = imread ('HI00008918.png')
```

```python
placa.min(),placa.max(),placa.dtype
```

```python
plt.imshow(placa,cmap='gray')
```

```python
placa2 = placa.astype('int') * 10
placa2[placa2>255] = 255
```

```python
placa2.max()
```

```python
plt.imshow(placa2,cmap='gray')
```

### Stretch de histograma

```python
placa2 = placa.copy().astype('int')
placa2 = placa2 - placa2.min()
placa2 = placa2 / placa2.max() * 255
placa2 = placa2.astype ('int')
```

```python
placa2.min(),placa2.max()
```

```python
plt.imshow(placa2,cmap='gray')
```
